```python
import pandas as pd
import glob
import os

# Pas dit aan naar waar jouw gedownloade bestanden staan
DATA_FOLDER = '.'
OUTPUT_FILE = 'raw_combined_data.csv'

print("Files Zoeken in map:", DATA_FOLDER)

# Alle CSV's zoeken (demanddata_2001.csv, etc.)
all_files = glob.glob(os.path.join(DATA_FOLDER, "demanddata_*.csv"))

if not all_files:
    print("Geen bestanden gevonden! Zorg dat de CSV's in deze map staan.")
else:
    df_list = [pd.read_csv(f) for f in all_files]
    raw_df = pd.concat(df_list, ignore_index=True)
    print(f"Totaal {len(raw_df)} rijen ingeladen.")

    raw_df.to_csv(OUTPUT_FILE, index=False)
    print(f"Bestand opgeslagen als: {OUTPUT_FILE}")
```
