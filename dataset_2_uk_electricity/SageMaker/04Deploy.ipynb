```python
import pandas as pd
import pickle

# 1. Laad het opgeslagen model
print("Model laden...")
try:
    with open("electricity_model.pkl", "rb") as f:
        model = pickle.load(f)
    print("‚úÖ Model succesvol geladen!")
except FileNotFoundError:
    print("‚ùå Fout: 'electricity_model.pkl' niet gevonden. Draai eerst notebook 03_predict!")

# 2. De interactieve applicatie loop
print("\n--- ‚ö° ELECTRICITY PREDICTOR ‚ö° ---")
print("Vul de gegevens in. Typ 'stop' bij Tijdstip om te stoppen.\n")

while True:
    try:
        # Vraag input aan de gebruiker
        p_input = input("Tijdstip (Settlement Period 1-48) [of 'stop']: ")

        if p_input.lower() == 'stop':
            print("Applicatie gestopt. Tot ziens!")
            break

        period = int(p_input)
        month = int(input("Maand (1-12): "))
        day = int(input("Dag van de week (0=Maandag, 6=Zondag): "))

        # Maak DataFrame voor voorspelling
        input_data = pd.DataFrame({
            "settlement_period": [period],
            "month": [month],
            "day_of_week": [day]
        })

        # Voorspel
        pred = model.predict(input_data)[0]

        # Toon resultaat
        print(f"üîÆ Voorspelde Vraag: {int(pred)} MW")
        print("-" * 30)

    except ValueError:
        print("‚ö†Ô∏è Ongeldige invoer! Gebruik alleen hele getallen.")
    except Exception as e:
        print(f"‚ö†Ô∏è Er ging iets mis: {e}")
```
